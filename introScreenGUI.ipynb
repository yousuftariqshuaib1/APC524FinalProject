```python
import PySimpleGUI as sg
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from PIL import Image, ImageTk
import numpy as np

# Helper method to draw figure from PysimpleGUI Demo:
# https://github.com/PySimpleGUI/PySimpleGUI/blob/master/DemoPrograms/Demo_Pyplot_Bar_Chart.py
def draw_figure(canvas, figure):
    figure_canvas_agg = FigureCanvasTkAgg(figure, canvas)
    figure_canvas_agg.draw()
    figure_canvas_agg.get_tk_widget().pack()
    return figure_canvas_agg

def fig_maker_2D():
    plt.scatter(np.random.rand(1,10),np.random.rand(1,10))
    return plt.gcf()

def fig_maker_3D():
    plt.scatter(np.random.rand(1,10),np.random.rand(1,10))
    return plt.gcf()


first_column = [
    [sg.Canvas(key='3DPlot')],
    [sg.Button('update')],
    [sg.Image(key='-IMAGE2-')]
]

second_column = [
    [sg.Canvas(key='2DPlot')],
    [sg.HorizontalSeparator()],
    [sg.Image(key='-IMAGE4-')]
]

layout = [
    [
        sg.Column(first_column),
        sg.VSeparator(),
        sg.Column(second_column),
    ]
]

window = sg.Window("Image Viewer", layout, finalize=True)

im = Image.open('dumbshit.jpg')
image = ImageTk.PhotoImage(image=im)


window['-IMAGE2-'].update(data=image)
window['-IMAGE4-'].update(data=image)

# Make synthetic data

fig1 = fig_maker_3D()
fig2 = fig_maker_2D()
draw_figure(window["3DPlot"].TKCanvas, fig1)
draw_figure(window["2DPlot"].TKCanvas, fig2)
plt.close()

while True:
    event, values = window.read()
    if event == "update":
        fig1 = fig_maker_3D()
        fig2 = fig_maker_2D()
        draw_figure(window["3DPlot"].TKCanvas, fig1)
        draw_figure(window["2DPlot"].TKCanvas, fig2)
        plt.close()
        
    if event == sg.WIN_CLOSED:
        break

window.close()
```

```python
import PySimpleGUI as sg
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from PIL import Image, ImageTk


_VARS = {'window': False,
         'fig_agg1': False,
         'fig_agg2': False,
         'pltFig': False}


def draw_figure(canvas, figure):
    figure_canvas_agg = FigureCanvasTkAgg(figure, canvas)
    figure_canvas_agg.draw()
    figure_canvas_agg.get_tk_widget().pack(side='top', fill='both', expand=1)
    return figure_canvas_agg

first_column = [
    [sg.Canvas(key='3DPlot')],
    [sg.HorizontalSeparator()],
    [sg.Image(key='-IMAGE2-')]
]

second_column = [
    [sg.Image(key='-IMAGE3-')],
    [sg.HorizontalSeparator()],
    [sg.Image(key='-IMAGE4-')]
]

layout = [
    [
        sg.Column(first_column),
        sg.VSeparator(),
        sg.Column(second_column),
    ]
]

_VARS['window'] = sg.Window('Such Window',
                            layout,
                            finalize=True,
                            resizable=True,
                            location=(100, 100),
                            element_justification="right")

def drawChart():
    _VARS['pltFig'] = plt.figure()
    plt.scatter(np.random.rand(1,10),np.random.rand(1,10))
    _VARS['fig_agg'] = draw_figure(
        _VARS['window']['3DPlot'].TKCanvas, _VARS['pltFig'])


# Recreate Synthetic data, clear existing figre and redraw plot.

def updateChart():
    _VARS['fig_agg'].get_tk_widget().forget()
    plt.clf()
    plt.scatter(np.random.rand(1,10),np.random.rand(1,10))
    _VARS['fig_agg'] = draw_figure(
        _VARS['window']['3DPlot'].TKCanvas, _VARS['pltFig'])

# \\  -------- PYPLOT -------- //


drawChart()

im = Image.open('dumbshit.jpg')
image = ImageTk.PhotoImage(image=im)


_VARS['window']['-IMAGE2-'].update(data=image)
_VARS['window']['-IMAGE3-'].update(data=image)
_VARS['window']['-IMAGE4-'].update(data=image)


# MAIN LOOP
while True:
    event, values = _VARS['window'].read(timeout=200)
    if event == sg.WIN_CLOSED or event == 'Exit':
        break
    updateChart()

plt.close()
_VARS['window'].close()
```
